```python
def get_col_with_zero_std_dev(df:pd.DataFrame)->list:
    cols_to_drop=list()
    num_cols=[f for f in df.columns if df[f].dtype!='O']
    for col in num_cols:
        if df[col].std()==0:
            cols_to_drop.append(col)
    return cols_to_drop
```

```python
def redundant_cols(df:pd.DataFrame,missing_thresh=0.7)->list:
    cols_to_drop=list()
    cols_missing_ratio=df.isna().sum().div(df.shape[0])
    cols_to_drop=list(cols_missing_ratio[cols_missing_ratio>missing_thresh].index)
    return cols_to_drop
```

```python
cols_to_drop1=get_col_with_zero_std_dev(df)
print(cols_to_drop1)
```

```python
cols_to_drop2=redundant_cols(df)
print(cols_to_drop2)
```

```python
cols_to_drop=cols_to_drop2+cols_to_drop1
# cols_to_drop.append("Unnamed:0")
```

```python
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler

```

```python
imputer=KNNImputer(n_neighbors=3)
preprocessing_pipeline=Pipeline(
    steps=[
        ("imputer",imputer),
        ("scaler",RobustScaler())
    ]
)
preprocessing_pipeline
```

## Resampling of the data

```python
from sklearn.utils import resample
```

```python
df
```

```python
df=preprocessing_pipeline.fit_transform(df)
```

```python
df=pd.DataFrame(data=df,columns=preprocessing_pipeline.get_feature_names_out())
```

```python
df
```

```python
minority_class=df[df['Good/Bad']==2]
```

```python
majority_class=df[df['Good/Bad']==0]
```

```python
minority_class_resampled=resample(minority_class,replace=True,n_samples=len(majority_class),random_state=15)
```

```python
minority_class_resampled.shape
```

```python
majority_class.shape
```

```python
df_resampled=pd.concat([majority_class,minority_class_resampled])
```

```python
df_resampled
```

```python
target=df_resampled['Good/Bad']
df_resampled=df_resampled.drop("Good/Bad",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_resampled,target,test_size=0.15,random_state=85)
```

## Model Building and Training

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
models={
    "random forest":RandomForestClassifier(),
    "gradient":GradientBoostingClassifier(),
    "ada":AdaBoostClassifier(),
    # "knn":KNeighborsClassifier(),
    "logistic":LogisticRegression(),
    "tree":DecisionTreeClassifier()
}
```

```python
from sklearn.metrics import accuracy_score
```

```python
model=models['logistic']
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
def evaluate_models(models:dict):
    report=dict()
    best_model=None
    best_score=-1
    for name,model in models.items():
        m=model.fit(x_train,y_train)
        y_pred=m.predict(x_test)
        acc=accuracy_score(y_test,y_pred)
        if acc>best_score:
            best_score=acc
            best_model=name
        report[name]=acc
    return (report,best_model)
```

```python
report,best=evaluate_models(models)
```
